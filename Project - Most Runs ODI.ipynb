```python
# 9. Which teams have the most representation in the dataset?
```

```python
df["Teams"].unique()
```

```python
# List of team codes we are interested in
main_teams = ['IND', 'SL', 'AUS', 'PAK', 'SA', 'WI', 'NZ', 'BAN', 'ENG', 'IRE', 'ZIM']

# Function to determine the main team from the string
def get_main_team(team):
    for code in main_teams:
        if code in team:  # Check if the main team code is a substring
            return code
    return team  # Return the original value if no match is found

# Apply the function to create the 'Main_Team' column
df['Main_Team'] = df['Teams'].apply(get_main_team)
```

```python
# Counting team representation
team_counts = df['Main_Team'].value_counts()

# Bar plot of team representation
plt.figure(figsize=(8, 6))
sns.barplot(x=team_counts.index, y=team_counts.values, palette='Paired')
plt.title('Team Representation in the Dataset')
plt.xlabel('Main_Team')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Which players have the highest proportion of 'Fours' and 'Sixes' compared to their total runs?
```

```python
# Calculating proportion of Fours and Sixes
df['Fours_Proportion'] = (df['Fours'] * 4) / df['Runs']
df['Sixes_Proportion'] = (df['Sixes'] * 6) / df['Runs']

# Selecting top 10 players based on Fours_Proportion
top_10_fours = df.nlargest(10, 'Fours_Proportion')

# Selecting top 10 players based on Sixes_Proportion
top_10_sixes = df.nlargest(10, 'Sixes_Proportion')

# Plotting Fours and Sixes Proportion for Top 10 Players
plt.figure(figsize=(14, 7))

# Plot for Fours Proportion
plt.subplot(1, 2, 1)
sns.barplot(x='Player', y='Fours_Proportion', data=top_10_fours, palette='BuGn')
plt.title('Top 10 Players: Proportion of Runs from Fours')
plt.xlabel('Player')
plt.ylabel('Proportion of Total Runs')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees

# Plot for Sixes Proportion
plt.subplot(1, 2, 2)
sns.barplot(x='Player', y='Sixes_Proportion', data=top_10_sixes, palette='OrRd')
plt.title('Top 10 Players: Proportion of Runs from Sixes')
plt.xlabel('Player')
plt.ylabel('Proportion of Total Runs')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees

plt.tight_layout()
plt.show()
```
